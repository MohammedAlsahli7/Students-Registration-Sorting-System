```python
def genTotal(file): 
    #the function takes a file that has ids, names, abcenses and mid, classwork, final grades
    #and compute the totalmark of each student and then generates a new file with name, id and totalmark.  
    print("Generationg the total mark...")
    
    inputFile = open(file)
    outPutFile = open("TotalMark.txt","w") #to generate a new file that will have name, id and totalmarks.
    
    for line in inputFile:
        #convert every line to a list of id, name, abcences and mid, cloasswork and final grades.
        newLine = [] 
        line = line.strip()
        line = line.split(":")
        
        #compute the totalmark and then write the name, id and totalmark into the file.
        totalMark = float(line[3])+float(line[4])+float(line[5]) #totalmark=mid+classwork+final
        newLine.append(line[0]) #to append name
        newLine.append(line[1]) #to append ID
        newLine.append(totalMark) #to append classwork
        outPutFile.write(newLine[1]+":"+newLine[0]+":"+str(newLine[2])+"\n") # arranged it ("name:id:totalmark")

    print("*"*50)
    print("-"*21 + " Done!! " + "-"*21)
    print("*"*50)
    inputFile.close()
    outPutFile.close()
    print("Returning to the main list...")
    return main(file)
    
def add(file):
    print("Adding a student...")
    inputFile = open(file,"a")
    stdId = input("Enter the student ID: ")
    stdName = input("Enter the student name: ")
    abcs = input("Enter the student absences number: ")
    midGrade = input("Enter the student midterm exam grade out of 35%: ")
    classGrade = input("Enter the student classwork grade out of 25%: ")
    finalGrade = input("Enter the student final exam grade out of 40%: ")
    
    stdId=validate(validate(stdId,"integer"),"id") ; abcs = validate(abcs,"integer") ; midGrade=validate(validate(midGrade,"float"),"midterm")
    classGrade=validate(validate(classGrade,"float"),"classwork") ; finalGrade = validate(validate(finalGrade,"float"),"final")
    
    if (stdId and abcs and midGrade and classGrade and finalGrade) != None:
        inputFile.write("\n"+str(stdId)+":"+str(stdName)+":"+str(abcs)+":"+str(midGrade)+":"+str(classGrade)+":"+str(finalGrade))
        inputFile.close()
        print("*"*50)
        print("-"*21 + " Done!! " + "-"*21)
        print("*"*50)
        
        choice = input("Do you want to redo Generating the total mark? (y) for yes or anything for no: ")
        if choice in "yY":
            inputFile.close()
            return genTotal(file)
        
        print("Returning to the main list...")
        inputFile.close()
        return main(file)
    else:
        print("Error, Invalid Input! Please try again.")
        print("Returning to the main list...")
        inputFile.close()
        return main(file)
    
    
    
def remove(file):
    print("Removing a student...")
    inputFile = open(file)
    studentsById = {}
    studentsByName = {}
    for line in inputFile:
        lines = []
        line = line.strip()
        line = line.split(":")
        for i in line:
            lines.append(i)
        studentsById[lines[0]] = lines[0:6]
        studentsByName[lines[1]] = lines[0:6]
    inputFile.close()

    print("-"*50)
    print("a. By ID ")
    print("b. By Student Name ")
    print("* Caution, when using the name to delete a student, all student with the same name will be removed")
    print("-"*50)
    by = input("Your choice is: ")
    

    if by == "a":
        stdId = input("Enter the student's ID: ")
        if stdId in studentsById :
            print(stdId+"'s information are:", "ID: "+ studentsById[stdId][0] + ", Name: "+studentsById[stdId][1]+", Absences: "+studentsById[stdId][2]+", Midterm: "+studentsById[stdId][3]+", Classwork: "+studentsById[stdId][4]+", Final: "+studentsById[stdId][5])
            sure = input("* Are you sure you want to delet him? (y) for yes or anything for no: ")
            if sure == "y":
                studentsById.pop(stdId)
                inputFile = open(file,"w")
                values = studentsById.values()
                print(val)
                index = 0
                for i in values:
                    for j in i:
                        inputFile.write(j)
                        if i.index(j) != (len(i)-1):
                            inputFile.write(":")
                    if index != (len(values)-1):
                        inputFile.write("\n")
                        index+=1
                inputFile.close()
            else:
                print("You choosed to not remove him.")
                print("*"*50)
                print("-"*21 + " Done!! " + "-"*21)
                print("*"*50)
                print("Returning to the main list...")
                return main(file)
        else:
            print("Error, The entered ID was not found.")
            print("Returning to the main list...")
            return main(file)
            
    elif by == "b":
        
        stdName = input("Enter the student's name: ")
        if stdName in studentsByName :
            print(stdName+"'s information are:", "ID: "+ studentsByName[stdName][0] + ", Name: "+studentsByName[stdName][1]+", Absences: "+studentsByName[stdName][2]+", Midterm: "+studentsByName[stdName][3]+", Classwork: "+studentsByName[stdName][4]+", Final: "+studentsByName[stdName][5])
            print("* Caution, if this information is not for the student you want please retry with his ID")
            sure = input("Are you sure you want to delet him? (y) for yes or anything for no: ")
            if sure == "y":
                studentsByName.pop(stdName)
                inputFile = open(file,"w")
                values = studentsByName.values()
                index = 0
                for i in values:
                    for j in i:
                        inputFile.write(j)
                        if i.index(j) != (len(i)-1):
                            inputFile.write(":")
                    if index != (len(values)-1):
                        inputFile.write("\n")
                        index+=1
                inputFile.close()
            else:
                print("You choosed to not remove him.") 
                print("*"*50)
                print("-"*21 + " Done!! " + "-"*21)
                print("*"*50)
                print("Returning to the main list...")
                return main(file)
        else:
            print("Error, The entered name was not found.")
            print("Returning to the main list...")
            return main(file)
    else:
        print("you should enter a or b only.")
        print("Returning to the main list...")
        return main(file)
        
    print("*"*50)
    print("-"*21 + " Done!! " + "-"*21)
    print("*"*50)
    
    choice = input("Do you want to redo Generating the total mark? (y) for yes or anything for no: ")
    if choice in "yY":
        return genTotal(file)
    
    print("Returning to the main list...")
    return main(file)

def modifying(file):
    print("Modifying student's information...")
    #modifiyng
    #file = "Student1.txt"
    inputFile = open(file)
    students = {}
    for line in inputFile:
        lines = []
        line = line.strip()
        line = line.split(":")
        for i in line:
            lines.append(i)
        students[lines[0]] = lines[0:6]
    inputFile.close()
    stdId = input("Enter the student id that you want to Modify his info: ")

    if stdId in students :#the input file
        print("-"*50)
        print("a. Change the ID")
        print("b. Change the Name")
        print("c. Change the Absences")
        print("d. Change the Midterm grade")
        print("e. Change the Class work")
        print("f. Change the Final grade")
        print("-"*50)


        choice =input("Enter letter from a to f :").lower()
        if choice == "a" or choice == "b" or choice == "c" or choice == "d" or choice == "e" or choice == "f":

            if choice == "a":
                newId = validate(validate(input("Enter the new ID: "),"integer"),"id")
                if newId != None:
                    students[newId] = students[stdId]
                    students.pop(oldid)
                    students[newId].pop(0)
                    students[newid].insert(0,newId)
                else: print("Error, the ID should be 9 digits only") ; return main(file)

            elif choice == "b":
                newName = input("Enter the new name: ")
                students[stdId].pop(1)
                students[stdId].insert(1,newName)

            elif choice == "c":
                newAbsences = validate(input("Enter the new number of Absences :"),"integer")
                if newAbsences != None: students[stdId].pop(2) ; students[stdId].insert(2,newAbsences)
                else: print("Error, the input should be integer only") ; return main(file)
                    
            elif choice == "d":
                newMidterm = validate(validate(input("Enter the new Midterm score out of 35% :"),"float"),"midterm")
                if newMidterm != None: students[stdId].pop(3) ; students[stdId].insert(3,newMidterm)
                else: print("Error, the Input should be a float or an integer between 0 and 35"); print("Returning to the main list...") ; return main(file)
                    
            elif choice == "e":
                newClassWork = validate(validate(input("Enter the new class work score out of 25% :"),"float"),"classwork")
                if newClassWork != None: students[stdId].pop(4) ; students[stdId].insert(4,newClassWork)
                else: print("Error, the Input should be a float or an integer between 0 and 25"); print("Returning to the main list...") ; return main(file)

            elif choice == "f":
                newFinal = validate(validate(input("Enter the new Final score out of 40% :"),"float"),"final")
                if newFinal != None: students[stdId].pop(5) ; students[stdId].insert(5,newFinal)
                else: print("Error, the Input should be a float or an integer between 0 and 40"); print("Returning to the main list...") ; return main(file)

            inputFile = open(file,"w")
            values = students.values()
            index = 0
            for i in values:
                for j in i:
                    inputFile.write(str(j))
                    if i.index(j) != (len(i)-1):
                        inputFile.write(":")
                if index != (len(values)-1):
                    inputFile.write("\n")
                    index+=1
            inputFile.close()
            print("*"*50)
            print("-"*21 + " Done!! " + "-"*21)
            print("*"*50)
            
            choice = input("Do you want to redo Generating the total mark? (y) for yes or anything for no: ")
            if choice in "yY":
                return genTotal(file)
            
            print("Returning to the main list...")
            return main(file)
        
        else:
            print("Error, you should enter letter from a to f")
            print("Returning to the main list...")
            return main(file)
    else:
        print("No id like this in file")
        print("Returning to the main list...")
        return main(file)

    
def show(file):
    print("Showing a specific student...")
    inputFile = open(file)
    students = {}
    for line in inputFile:
        lines = []
        line = line.strip()
        line = line.split(":")
        for i in line:
            lines.append(i)
        students[lines[0]] = lines[0:6]
    inputFile.close()

    stdId = input("Enter the student id that you want to Modify his info: ")
    if stdId in students:
        print("-"*50)
        print(stdId+"'s information are:",
              "\nID: "+ students[stdId][0] + "\nName: "+students[stdId][1]+ 
              "\nAbsences: "+students[stdId][2] + "\nMidterm: "+students[stdId][3]+
              "\nClasswork: "+students[stdId][4] + "\nFinal: "+students[stdId][5])
        print("*"*50)
        print("-"*21 + " Done!! " + "-"*21)
        print("*"*50)
        print("Returning to the main list...")
        return main(file)
    else:
        print("No id like this in file")
        print("Returning to the main list...")
        return main(file)
    
def criteria(file):
    print("Searching by a criteria...")
    inputFile = open(file)
    students = {}
    for line in inputFile:
        lines = []
        line = line.strip()
        line = line.split(":")
        for i in line:
            lines.append(i)
        students[lines[0]] = lines[0:6]
    inputFile.close()

    print("-"*50)
    print("a. Search by the Absences")
    print("b. Search by the Midterm grade")
    print("c. Search by the Class work")
    print("d. Search by the Final grade")
    print("-"*50)
    
    criteria = input("Enter the criteria (from a to d): ").lower()
    
    if criteria == "a": criteria = 2
    elif criteria == "b": criteria = 3
    elif criteria == "c": criteria = 4
    elif criteria == "d": criteria = 5
    else: 
        print("Error, the input should be (from a to d)")
        print("Returning to the main list...")
        return main(file)
    
    
    sign = input("Enter the the case sign ( < , > , = ): ")
    if sign != "<" and sign != ">" and sign != "=":
        print("Error, wrong input!")
        print("Returning to the main list...")
        return main(file)
    
    
    condition = validate(input("Enter the number: "),"float")
    if condition == None:
        print("Error, wrong input!")
        print("Returning to the main list...")
        return main(file)
    

        
    dic_values = students.values()
    values = []
    for lists in dic_values:
        values.append(lists)
    flag = 0

    num = 1
    if sign == "<":
        print("the information of the students who met the criteria are:")
        for lisT in values:
            if float(lisT[criteria]) < condition:
                print(num, end=" - ")
                num +=1
                print("ID: "+ lisT[0] + ", Name: "+lisT[1]+", Absences: "+lisT[2]+", Midterm: "+lisT[3]+", Classwork: "+lisT[4]+", Final: "+lisT[5])
            else: flag +=1
    elif sign == ">":
        print("the information of the students who met the criteria are:")
        for lisT in values:
            if float(lisT[criteria]) > condition:
                print(num, end="- ")
                num +=1
                print("ID: "+ lisT[0] + ", Name: "+lisT[1]+", Absences: "+lisT[2]+", Midterm: "+lisT[3]+", Classwork: "+lisT[4]+", Final: "+lisT[5])
            else: flag +=1
    elif sign == "=":
        print("the information of the students who met the criteria are:")
        for lisT in values:
            if float(lisT[criteria]) == condition:
                print(num, end="- ")
                num +=1
                print("ID: "+ lisT[0] + ", Name: "+lisT[1]+", Absences: "+lisT[2]+", Midterm: "+lisT[3]+", Classwork: "+lisT[4]+", Final: "+lisT[5])
            else: flag +=1
    if flag == len(values):
        print("There's none.")
    print("*"*50)
    print("-"*21 + " Done!! " + "-"*21)
    print("*"*50)
    print("Returning to the main list...")
    main(file)


def validate(objects , validation):
    # This function is about validationits, contains all the conditions needed.
    try:
        if objects != None :
        
            if validation == "main" :
                objects = int(objects)
                if objects < 1 or objects > 8:
                    raise ValueError
                return objects

            elif validation == "file" :
                file = open(objects)
                x= 0
                for line in file:
                    if len(line.split(":")) == 6:
                        line = line.strip()
                        line = line.split(":")
                        line[3] = validate(line[3],"float") ; line[4] = validate(line[4],"float") ; line[5] = validate(line[5],"float") ; line[2] = validate(line[2],"integer")
                        line[3] = validate(line[3],"midterm") ; line[4] = validate(line[4],"classwork") ; line[5] = validate(line[5],"final")

                    if len(line) < 6 or len(line[0]) != 9 or len(str(line[2])) == 0 or len(str(line[2])) > 2 or ( (line[3] or line[4] or line[5]) == None ):
                        x+=1
                    if x > 0:
                        print("Error, Invalid file format! Please try again.")
                        return None
                
                file.close()
                return objects

            elif validation == "integer" :
                objects = int(objects)
                return objects

            elif validation == "float" :
                objects = float(objects)
                return objects

            elif validation == "midterm" :
                if objects < 0 or objects > 35:
                    raise ValueError
                return objects

            elif validation == "classwork" :
                if objects < 0 or objects > 25 :
                    raise ValueError
                return objects

            elif validation == "final":
                if objects < 0 or objects > 40 :
                    raise ValueError
                return objects

            elif validation == "id":
                if len(str(objects)) != 9 :
                    raise ValueError   
                return objects


    except ValueError:
        return None
        
    except IOError :
        print("Error, No such file was found!")
        return None

def main(inputFile):
    #the main menu
    students = inputFile  #to open the file that the user write ("name:id:Absences:Midterm:Classwork:Final")
    choice = validate(input("Enter a number from the list above (1-7) or 8 to reprint the list: "),"main")
    while choice == None:
        print("Invalid Input, the input should be a number between 1 and 8, please try again.") 
        choice = validate(input("Enter a number from the list above: "),"main") #for the user to enter a number from 1 to 7 from the menu
    if choice == 1:
        genTotal(students)
    elif choice == 2:
        criteria(students)
    elif choice == 3:
        add(students)
    elif choice == 4:
        remove(students)
    elif choice == 5:
        modifying(students)
    elif choice == 6:
        show(students)
    elif choice == 7:
        print("*"*50)
        print("---The program has been terminated successfully---")
        print("*"*50)
        return None
    elif choice == 8:
        start(inputFile)
        
# Start the program
def start(file):  
    print("*"*50)
    print("1-Generate a new file that has Name, ID, and the total mark")
    print("2-Show students that meet a specific criteria")
    print("*3-Adding a Student to the class")
    print("*4-Removing a student from the class")
    print("*5-Modifying a student information")
    print("6-Show the data of a student")
    print("7-Terminating a program")
    print("8-Reprinting this list")
    print("*"*50)
    print()
    main(file)

    
def file():
    #inputFile = validate(input("Enter a file name: "),"file")
    #while inputFile == None :
    #    inputFile = validate(input("Enter a file name: "),"file")
    inputFile = "LabProject.txt"    
    start(inputFile)
# Start asking for the name of the file LabProject.txt
file()
```
